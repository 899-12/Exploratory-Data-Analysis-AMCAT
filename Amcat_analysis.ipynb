## 2.Outlier Detection and Descriptive Statistics

```python
# Function to detect outliers
def find_outliers(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[col] < lower_bound) | (df[col] > upper_bound)]

# Get outliers for each numerical column
for col in numerical_columns:
    outliers = find_outliers(col)
    print(f'Outliers in {col}:')
    print(outliers)

# Calculate statistics for each numerical column
for col in numerical_columns:
    print(f'{col} - Min: {df[col].min()}, Max: {df[col].max()}, Median: {df[col].median()}, '
          f'Std: {df[col].std()}, Kurt: {df[col].kurtosis()}, Skew: {df[col].skew()}')
```

##### 3.Univariate Analysis for Categorical Columns

```python
categorical_columns = ['Designation', 'JobCity', 'Gender', 'Degree', 'Specialization']


def reduce_categories(df, col, top_n=5):
    # Get the top n categories
    top_categories = df[col].value_counts().nlargest(top_n).index
    # Create a new column with reduced categories
    df[col + '_reduced'] = df[col].where(df[col].isin(top_categories), 'Other')
    return col + '_reduced'


for col in ['Designation', 'JobCity', 'Specialization']:
    reduce_categories(df, col)

plt.figure(figsize=(12, 8))
for i, col in enumerate(categorical_columns):
    # Use the reduced category columns where applicable
    x_col = col + '_reduced' if col in ['Designation', 'JobCity', 'Specialization'] else col
    plt.subplot(2, 3, i + 1)
    sns.countplot(data=df, x=x_col)
    plt.title(f'Countplot of {x_col}')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### Observation
#### Most of the respondents are males, primarily holding a B.Tech/B.E. degree, working in varied designations, with Bangalore being a common city. Specializations lean towards technical fields like computer engineering and information technology.

### Gender and Specialization Analysis

```python
from scipy.stats import chi2_contingency


specialization_columns = ['ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg']  
contingency_table = pd.crosstab(df['Gender'], df[specialization_columns].sum(axis=1))

chi2, p, _, _ = chi2_contingency(contingency_table)

print(f'Chi-Square Statistic: {chi2}, p-value: {p}')

if p < 0.05:
    print("There is a significant relationship between gender and specialization.")
else:
    print("There is no significant relationship between gender and specialization.")
```
